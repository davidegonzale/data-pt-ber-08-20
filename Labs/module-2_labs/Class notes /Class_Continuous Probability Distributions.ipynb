# Continuos Distributions

### Continuos Uniform Distribution

```python
import numpy as np
import pandas as pd 
import scipy.stats
import matplotlib.pyplot as plt 
from scipy.stats import uniform
%matplotlib inline
```

```python
cuni = uniform()
test= cuni.rvs(size=10)
print(test)
print(test[0])
```

```python
x = cuni.rvs(size=1000)
plt.hist(x, bins=100)
```

```python
#np.linspace
numrange = np.linspace(-0.1, 1.1, 100)
plt.plot(numrange, cuni.pdf(numrange))
```

```python
numrange2 = np.linspace(-0.1, 1.1, 100)
plt.plot(numrange2, cuni.cdf(numrange2))
```

```python
cuni = uniform(0, 10)
numrange = np.linspace(-1, 11, 100)
plt.plot(numrange, cuni.pdf(numrange))
```

```python
from scipy.stats import norm
```

```python
normal = norm(0,1)
numerange = np.linspace (-5,5, 100)
plt.plot(numerange, normal.pdf(numerange))
```

```python
plt.plot(numerange, normal.cdf(numerange))
```

```python
x = normal.rvs(size=10000)

plt.hist(x, bins=100)
plt.show()
```
