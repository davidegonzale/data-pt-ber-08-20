```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd 
import seaborn as sns
sns.set_palette("husl")

import warnings
warnings.filterwarnings('ignore')
```

# Seaborn

```python
df = pd.DataFrame()
df['x'] = np.arange(0,10,1)
df['y'] = df['x']**2
print(df)
```

```python
#plt.plot(df['x'], df['y'])
df.plot('x','y')
```

```python
sns.lineplot(df['x'], df['y'])
```

```python
sns.lmplot('x', 'y', data=df)
```
